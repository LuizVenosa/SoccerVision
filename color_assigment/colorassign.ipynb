```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image_path = '../output_videos/player_2.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```

```python
#top half of the image

top_half_img = image[0: int(image.shape[0]/2), :]
plt.imshow(top_half_img)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Reshape the top_half_img to a 2D array of pixels
pixels = top_half_img.reshape(-1, 3)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(pixels)

# Replace each pixel value with its corresponding cluster center
clustered_img = kmeans.cluster_centers_[kmeans.labels_].reshape(top_half_img.shape).astype('uint8')

# Display the clustered image
plt.imshow(clustered_img)
plt.show()
```
